```python
import pandas as pd
import numpy as np
import random
import time
import math
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt
import pickle
```

```python
df2 = pd.read_csv("db_data.csv")
df2
```

```python
# Read data and pre-process
df = pd.read_csv("mockdata.csv")
df['basePrice'] = df.apply(lambda row: float(row['basePrice'].strip()[1:]), axis=1)
df['quantity'] = np.random.randint(1,50,df.shape[0])
df['wastage'] = np.random.randint(0,50, df.shape[0])/100
df2 = pd.read_csv("db_data.csv")
df = pd.concat([df, df2])
df['dynamicPrice'] = df.apply(lambda row: row['basePrice'] - 0.6*row['basePrice']/(1+math.log(row['daysToExpiry'])), axis=1)
```

```python
# Instatiate Random Forest model and fit
model = RandomForestRegressor(random_state=1, n_estimators=10)
x = df.iloc[:, 1:-1].values
y = df.iloc[:, -1].values
model.fit(x[100:], y[100:])

# Make predictions
pred = model.predict(x)
```

```python
# View predictions
df['predictedPrice'] = pred
df = df.drop("dynamicPrice", axis=1)
df
```

```python
# Save model
filename = 'rf_model.sav'
pickle.dump(model, open(filename, 'wb'))
```
